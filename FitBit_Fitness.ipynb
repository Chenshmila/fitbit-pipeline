# Explanation of the data and selection of relevant tables

### **Daily Level Data**

* **dailyActivity:** The most comprehensive daily table. It includes total steps, total distance, a breakdown of activity levels (Very, Moderately, Lightly, and Sedentary) by both minutes and distance, and total calories burned.
* **dailyCalories:** Focuses strictly on the total number of calories burned by each user per day.
* **dailyIntensities:** Provides a detailed breakdown of the time and distance spent in each intensity category per day.
* **dailySteps:** Contains only the total step count for each user per day.
* **sleepDay:** Summarizes daily sleep data, including the number of sleep records, total minutes spent asleep, and total time spent in bed.

### **Hourly Level Data**

* **hourlyCalories:** The amount of calories burned during each specific hour of the day.
* **hourlyIntensities:** Shows the total and average activity intensity for each hour.
* **hourlySteps:** The total number of steps taken during each hour.

### **Minute Level Data**

These tables come in two formats: **Narrow** (one row per minute) and **Wide** (one row per hour with 60 columns representing each minute).

* **minuteCalories (Narrow/Wide):** Calories burned tracked at a minute-by-minute granularity.
* **minuteIntensities (Narrow/Wide):** The intensity level recorded for every minute.
* **minuteSteps (Narrow/Wide):** The exact number of steps taken in every minute.
* **minuteMETsNarrow:** Displays Metabolic Equivalent of Task (MET) values per minute, which represent the energy cost of physical activities.
* **minuteSleep:** Logs the sleep state (represented by numeric values for sleep stages/wakefulness) for every minute of a sleep log.

### **Additional Logs**

* **heartrate_seconds:** High-resolution heart rate data, recorded approximately every 5 seconds.
* **weightLogInfo:** User-logged weight data, including Weight (Kg/Pounds), BMI, Body Fat (rarely populated), and whether the report was submitted manually.

### Data cleaning

```python
data_dict = {
    "dailyActivity":     dailyActivity,
    "sleepDay":          sleepDay,
    "hourlySteps":       hourlySteps,
    "heartrate_seconds": heartrate_seconds,
    "weightLogInfo":     weightLogInfo
}
```

```python
def summarize_df(df, name):
    print("\n" + "="*40)
    print("Summary for", name)
    print("="*40)
    print("Shape:", df.shape)
    print("Columns:", list(df.columns))
    df.info()
    print(df.head())
```

```python
def basic_clean(df, columns_list):
    print("Total duplicated rows:", df.duplicated().sum())
    print(f"Duplicated based on {columns_list}:", df.duplicated(subset=columns_list).sum())
    print("Null values per column:\n", df.isnull().sum())
    
    return df
```

## Clean dailyActivity df

```python
summarize_df(dailyActivity, "dailyActivity")
```

```python
basic_clean(dailyActivity, ["Id", "ActivityDate"])
```

```python
dailyActivity["ActivityDate"] = pd.to_datetime(dailyActivity["ActivityDate"])
print(dailyActivity["ActivityDate"].dtype)
```

```python
dailyActivity = dailyActivity.rename(columns={
    "Id": "user_id",
    "ActivityDate": "activity_date",
    "TotalSteps": "total_steps",
    "Calories": "calories"
})
```

```python
dailyActivity[dailyActivity["total_steps"] == 0]
```

```python
dailyActivity[dailyActivity["calories"] == 0]
```

Mark rows with zero steps or calories as incomplete or potentially missing data
This helps to track data quality without deleting any records

```python
dailyActivity['incomplete'] = (dailyActivity['total_steps'] == 0) | (dailyActivity['calories'] == 0)
```

```python
dailyActivity["calories_per_step"] = dailyActivity["Calories"] / dailyActivity["TotalSteps"]
```

```python
dailyActivity.to_csv("clean_daily_activity.csv", index=False)
```

## Clean sleepDay Df

```python
summarize_df(sleepDay, "sleepDay")
```

```python
basic_clean(sleepDay, ["Id", "SleepDay"])
```

```python
sleepDay = sleepDay.drop_duplicates(subset=["Id","SleepDay"])
```

```python
sleepDay = sleepDay.copy()
sleepDay.loc[:, "SleepDay"] = pd.to_datetime(
    sleepDay["SleepDay"],
    format="%m/%d/%Y %I:%M:%S %p",
    errors='coerce'
)

print(sleepDay["SleepDay"].dtype)
print(sleepDay[sleepDay["SleepDay"].isna()]) 
print(type(sleepDay["SleepDay"].iloc[0]))
```

```python
sleepDay = sleepDay.rename(columns={
    "Id": "user_id",
    "SleepDay": "sleep_date",
    "TotalSleepRecords": "total_sleep_records",
    "TotalMinutesAsleep": "minutes_asleep",
    "TotalTimeInBed": "minutes_in_bed"
})
```

```python
sleepDay['minutes_awake_in_bed'] = sleepDay['minutes_in_bed'] - sleepDay['minutes_asleep']
```

```python
sleepDay['incomplete'] = (sleepDay['minutes_asleep'] == 0) | (sleepDay['minutes_in_bed'] == 0)
```

```python
sleepDay.to_csv("clean_sleep_day.csv", index=False)
```

## Clean weightLogInfo Df

```python
summarize_df(weightLogInfo, "weightLogInfo")
```

```python
basic_clean(weightLogInfo, ["Id", "Date"])
```

```python
# The 'Fat' column contains 65 missing values out of 67 rows in the dataset.
# Because most of the values are missing, the column provides very little useful information for analysis and could introduce noise or bias.
# Therefore, I remove it from the dataset.
weightLogInfo = weightLogInfo.drop(columns=["Fat"])
```

```python
# 'WeightPounds' represents the same information as 'WeightKg' but in a different unit. 
# Since 'WeightKg' is already available, keeping both columns is redundant, so we remove 'WeightPounds'.
weightLogInfo = weightLogInfo.drop(columns=["WeightPounds"])
```

```python
weightLogInfo['Date'] = pd.to_datetime(
    weightLogInfo['Date'], 
    format="%m/%d/%Y %I:%M:%S %p", 
    errors='coerce'
)
print(weightLogInfo["Date"].dtype)
```

```python
weightLogInfo['missing_weight'] = (weightLogInfo['WeightKg'] == 0)
```

```python
weightLogInfo = weightLogInfo.rename(columns={
    'Id': 'user_id',
    'Date': 'date',
    'WeightKg': 'weight_kg',
    'BMI': 'bmi',
    'IsManualReport': 'manual_report',
    'LogId': 'log_id'
})
```

```python
weightLogInfo.to_csv("weightLogInfo_clean.csv", index=False)
```
